# 13명이서 할 수 있는 간단한 교실게임 - 마피아

#### 2주차 미션으로 시도한 마피아 게임을 수정, 보완해서 우리반 학생들과 실제로 할 수 있는 마피아 게임 만들기

#### 수정보완할 점: 
1) 13명 인원수에 맞게 역할부여

2) 화면 깔끔하게 디자인(인물소개, 게임설명, 게임시작 메뉴 추가)


3) 게임종료조건이 달성되면 게임이 종료!

```python
import random
import time

#반 아이들 명단 불러오기+

data = open('학생명단.txt')

#마피아 게임 역할 정하기- 시민 8명(p1-p8), 수사관 1명(p9), 마피아 3명(m1-m3), 대부 1명(m4)
mafia = 4
civil = 0
names=[]
print(mafia)
print(civil)
for row in data :
  names.append(row.strip())


random.shuffle(names)
p1=names[0]
p2=names[1]
p3=names[2]
p4=names[3]
p5=names[4]
p6=names[5]
p7=names[6]
p8=names[8]
p9=names[9]
m1=names[10]
m2=names[11]
m3=names[12]
m4=names[13]
while True:
  print('=========================================================================================================')
  print(' <<마피아 게임>>')
  print('   메   뉴\n (1)게임역할소개 \n (2)게임설명 \n (3)게임시작')
  menu = input('메뉴를 보고 원하는 것을 입력해주세요: ')

  if menu == '게임시작':
    print('=======================================================================================================')
    print('마피아 게임을 시작합니다.')
    print('지금부터 마피아 역할 지정이 있겠습니다. 사회자만 다음 화면을 확인하기 바랍니다.')
    input('엔터를 누르면 마피아 역할 지정 결과를 확인할 수 있습니다. \n \n')


    print('======================사회자만 확인하세요!!!!!!!!!!!!!!!!!!!!!!!!!!!!!================================')
    print('시민 : ' + p1 + ', ' + p2 + ', ' + p3 + ', ' + p4 + ', ' + p5 + ', ' + p6 + ', ' + p7 + ', ' + p8)
    print('수사관 : ' + p9)
    print('마피아: '+ m1 + ', ' + m2 + ', ' + m3)
    print('대부:' + m4 )
    print('======================================================================================================')
    print('\n \n 사회자는 각 학생들에게 역할을 조용히 알려줍니다(쪽지로 건네주는 것을 추천함)')
    time.sleep(1)
    input('다알려주었으면 엔터를 눌러주세요!')


    print('\n \n \n지금부터 게임을 시작합니다. ')
    time.sleep(1)
    print('중요!: 반드시 사회자의 안내로 진행합니다. 사회자는 모든 플레이어의 캐릭터를 알고 있어야 합니다.\n \n \n')
    time.sleep(2)
    print('그럼 첫 번째 밤부터 시작합니다.')
    time.sleep(1)
    print('모두들 눈을 감아주세요!')
    time.sleep(1)
    print('밤입니다. 마피아 또는 대부 카드를 가진 플레이어는 모두 눈을 뜨고 서로의 정체를 확인합니다.')
    time.sleep(1)
    print('눈 감은 사람들이 눈치채지 못하게 살해할 대상을 3명을 순서대로 정해주세요.')
    print('앞으로 마피아 그룹에는 이 순서를 바꾸기 위해 논의할 시간이 따로 주어지지 않습니다.')
    print('만약 그들이 계획을 바꾸고 싶다면 반드시 낮 단계에서 다른 사람이 알아채지 못할 교모한 신호나 단서를 주어야 합니다.')
    print('이 모든 과정이 끝나거나 60초가 지나면, 이제 마피아 그룹은 눈을 감고 잠이 듭니다.')
    input('위의 글을 다 읽고 엔터를 누르면 60초세기가 시작됩니다.')

    for i in range(1):
      print(i+1)
      time.sleep(1)

    print('60초가 지났습니다. 마피아는 눈을 감으세요. 이제부터는 수사관의 조사가 있겠습니다.')
    time.sleep(1)
    print('수사관은 눈을 뜨고 마피아이거나 대부일 것 같은 사람을 지목하세요.')
    input('지목이 되었으면 사회자는 그 결과를 조용히 고갯짓으로 알려줍니다. 알려줬으면 엔터를 누르세요.')
    print('이제는 대부 차례입니다. 대부는 수사관일 것 같은 사람을 지목하세요.')
    input('지목이 되었으면 사회자는 그 결과를 조용히 고갯짓으로 알려줍니다. 알려줬으면 엔터를 누르세요.')
    print('밤단계가 끝났습니다. 이제는 낮이 시작되었습니다.')
    time.sleep(1)
    print('모든 플레이어는 눈을 뜨고, 1번부터 차례대로 마피아나 대부일 것 같은 사람을 지목합니다.')
    answer = input('가장 많이 지목된 사람을 입력해 주세요: ')


    if answer == m1 or answer == m2 or answer == m3 or answer == m4:
      print('네, ' + answer + '는 마피아입니다.')
    else:
      print('무고한 시민이 희생되었습니다.')
      print('=================================================================================================================')
      civil = civil +1
      

    while True:
      if civil == mafia:
       print('마피아가 승리하였습니다. 게임을 종료합니다.')
       break
      elif mafia == 0:
       print('시민이 승리하였습니다. 게임을 종료합니다.')
       break
      else:
       print('밤입니다. 마피아는 눈을 뜨지말고 살해할 시민을 한 명 투표합니다. 세 명이 일치하면 그 시민은 살해됩니다.')
       answer2 = input('만장일치가 되었다면 이름을 입력하고, 그렇지 않다면 엔터를 눌러주세요;')
       if answer2=="":
        print("")
       else:
        civil = civil + 1
        print(civil)
        print(answer2 + '가 살해되었습니다.')
        time.sleep(1)
       print('수사관은 눈을 뜨고 마피아이거나 대부일 것 같은 사람을 지목하세요.')
       input('지목이 되었으면 사회자는 그 결과를 조용히 고갯짓으로 알려줍니다. 알려줬으면 엔터를 누르세요.')
       print('이제는 대부 차례입니다. 대부는 수사관일 것 같은 사람을 지목하세요.')
       input('지목이 되었으면 사회자는 그 결과를 조용히 고갯짓으로 알려줍니다. 알려줬으면 엔터를 누르세요.')
       print('밤단계가 끝났습니다. 이제는 낮이 시작되었습니다.')
       time.sleep(1)
       print('모든 플레이어는 눈을 뜨고, 1번부터 차례대로 마피아나 대부일 것 같은 사람을 지목합니다.')
       answer = input('가장 많이 지목된 사람을 입력해 주세요: ')
       if answer == m1 or answer == m2 or answer == m3 or answer == m4:
        mafia  = mafia -1
        print('네, ' + answer + '는 마피아입니다.')
        print(mafia)
        print('=================================================================================================================')
       else:
        civil = civil + 1
        print(civil)        
        print('무고한 시민이 희생되었습니다.')
        print('=================================================================================================================')

  elif menu == '게임역할소개':
    print('=================================================================================================================')
    print('게임역할 소개를 시작합니다.')

    while True:
      role = input('시민, 수사관, 마피아, 대부 중 알고 싶은 역할을 입력해주세요(다 알았으면 <나가기> 입력):')
      if role == '시민':
        print('시민: 밤에는 잠이들고, 낮에는 토론을 하며 마피아나 대부를 찾아 처형해야함.')
      elif role == '수사관':
        print('수사관: 시민과 역할이 같으며 \n 그것과 더불어 밤마다 한 명의 플레이어를 지목해서 그 플레이어가 마피아인지 아닌지 확인할 수 있음.')
      elif role == '마피아':
        print('마피아: 범죄자 그룹이며 목표는 시민을 살해하는 것. 마피아 수가 시민의 수와 같거나 많아지면 게임에서 승리한다.')
      elif role == '대부':
        print('대부: 마피아그룹에서 가장 중요한 사람. 마피아와 같으며 밤마다 한 명을 지목해서 그사람이 수사관인지 아닌지 알 수 있다.')
      else:
        break
    
  elif menu == '게임설명':
    print('게임설명을 시작합니다.')
    time.sleep(2)
    print('모든 플레이어는 각자 자신의 캐릭터 카드를 확인합니다. \n')
    print('플레이어들은 게임 중 언제든지 자신의 캐릭터 카드를 확인할 수 있지만, \n이때 다른 사람에게 자신의 캐릭터가 보이지 않도록 주의해야합니다.')
    print('\n 게임은 밤 단계와 낮 단계를 번갈아가면서 진행합니다.')
    print('밤 단계에서는 범죄자 그룹에게 플레이어 1명을 살해할 기회가 주어지고, \n 낮 단계에서는 모든 플레이어가 토론을 통해 범죄자 그룹으로 의심되는 플레이어 1명을 처형할 수 있습니다.')
    print('*단 첫 번째 밤에는 아무도 살해당하지 않습니다.')
    print('\n 게임이 시작되면 사회자는 첫 번째 밤이 시작되었다고 선언합니다.')
    print('밤이 시작되면, 모든 플레이어는 반드시 눈을 감아야 합니다.')
    print('\n 밤 단계는 3개의 세부 단계로 구성되어 있습니다. 사회자는 각 세부단계에서 어떤 일이 벌어지는지 알려주고, \n 플레이어들이 어떻게 행동해야 할지 안내합니다.')
    print('\n *중요: 이때 플레이어들이 임의로 단계를 진행하지 않도록 주의하세요. 각 단계는 반드시 사회자의 안내로 진행되어야 합니다.')
    print('\n *참고: 사회자는 모든 플레이어의 캐릭터를 알고 있어야 합니다. 매번 캐릭터 카드를 확인하거나 기억해두기가 어렵다면, \n 자신만 볼 수 있게 따로 적어둘 것을 권장합니다.')
    
```
